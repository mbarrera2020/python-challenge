```python
# main -- for python-challenge PyBank
import os
import csv
```

```python
# setup path for input file location
csvpath=os.path.join('Resources','budget_data.csv')
csvpath
```

```python
# set variables to 0 -- this code is working!
months = 0
total = 0
diff = 0

# read the input file
with open(csvpath) as csvfile:
    csvreader = csv.reader(csvfile, delimiter=',')
    
    # determine the number of rows to read in the input file
    num_rows = len(list(csvreader))
    
    # reduce rows by 1 since first row is a header
    months = num_rows -1
    print("# of months: ", months)
```

```python
# set variables to 0 -- -- this code is working!
total = 0

with open(csvpath, newline='', encoding='utf-8') as inputfile:
    csvreader=csv.reader(inputfile, delimiter=',')
        
    next(inputfile)   # skip header row

    #Lists to store data
    date_lst=[]
    profitloss_lst=[]
    
    for row in csvreader:
            # add date to the list
            date_lst.append(row[0])                
            
            # convert string amount to integer
            pf=int(row[1])
            
            # add integer amount to the profit/loss list
            profitloss_lst.append(pf)
            
            # track total amount
            total = total + pf
             
    # TEMP code - to show if data got stored 
    print("The following are the contents of the date list:")
    print(date_lst)
    print()
    print("The following are the contents of the profit_loss list:")
    print(profitloss_lst)
    print()
    print("Total amount = ", total)     # Note:  expected result is 38,382,578
    print()
    
```

```python
# use the profit_loss list to create a change (aka amount difference) list -- this code is working!
# set variables to 0
count = 1
diff = 0
total_diff = 0

  
#List to store change in amount data
change_lst=[]

# temp code for testing / debugging only
#print(count)
#print(diff)
#print(months)
change_lst.append(0)

while count != months:
    # calculate the difference and store in change list
    diff = profitloss_lst[count] - profitloss_lst[count-1]
            
    # add change amount to the list
    change_lst.append(diff)
    
    # track total change amount
    total_diff = total_diff + diff
  
    count = count + 1
    
```

```python
# TEMP code - to show if data got stored in change list
print("The following are the contents of the change list:")
print(change_lst)
print()
# convert to dollar currency
total_dc = "${:,.2f}".format(total)
print("Total (expected result: $38,382,578) = ", total_dc)
print()

# average using total / # of items
avg1 = round(sum(change_lst) / len(change_lst),2)
# convert to dollar currency
avg1dc = "${:,.2f}".format(avg1)

print("Average change using total / # of items (expected result: $-2,288.20) = ", avg1dc)
print()

# need import to make MEAN function work
from statistics import mean

# average using mean average function
avg2 = round(mean(change_lst),2)
# convert to dollar currency
avg2dc = "${:,.2f}".format(avg2)

print("Average change using average mean function should be $-2,315.12 = ", avg2dc)
print()

greatest_inc = max(change_lst)
# convert to dollar currency
greatest_incdc = "${:,.2f}".format(greatest_inc)
print("Greatest Increase (expected result: $1,926,159): ", greatest_incdc)
print()
greatest_dec = min(change_lst)
# convert to dollar currency
greatest_decdc = "${:,.2f}".format(greatest_dec)
print("Greatest Decrease (expected result: $-2,196,167): ", greatest_decdc)
```

```python
#combine all lists as a clean revised budget csv & store the file in the <Resources> folder -- code is working!
import os
import csv 

cleaned_csv=zip(date_lst, profitloss_lst, change_lst)

output_file=os.path.join('Resources','budget_data_rev.csv')

with open(output_file, 'w', newline='') as datafile:
    writer=csv.writer(datafile)
    
    #writer.writerow(["Date", "Profit/Losses", "Differences"]) -- no headers
    writer.writerow(["Date", "Profit/Losses", "Differences"])
        
    writer.writerows(cleaned_csv)

# display output file to confirm it got created in the folder
print("The revised budget file has been created with a header row: ")
print(output_file)

# TEMP code - to show if data got stored in revbudget_lst list
#print("The following are the contents of the revised budget list:")
#print(cleaned_csv)
#print()
```
